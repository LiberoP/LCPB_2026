```python
import sys
import os
from mace.cli.run_train import main


os.makedirs("/kaggle/working/checkpoints", exist_ok=True) 
os.makedirs("/kaggle/working/results", exist_ok=True)
dataset_path = "/kaggle/input/datasets/niccoldepoli/my-mace-data"
custom_args = [
    "mace_run_training", 
    "--name=model_cristal",
    f"--train_file={dataset_path}/train.extxyz",
    f"--valid_file={dataset_path}/valid.extxyz",
    "--energy_key=energy",
    "--forces_key=forces",
    "--E0s=average",             
    "--model=MACE",
    "--num_interactions=2",
    "--max_num_epochs=50",
    "--batch_size=5",            
    "--device=cuda",
    "--default_dtype=float32",
    "--r_max=5.0",              
    "--checkpoints_dir=/kaggle/working/checkpoints",
    "--results_dir=/kaggle/working/results",
    "--keep_checkpoints"
]

sys.argv = custom_args

if __name__ == "__main__":
    try:
        main()
        print("Training completed!")
    except Exception as e:
        print(f"Error: {e}")
```
